```python
# A)

# Selection Sort corrigido
def selection_sort(arr):
    """
    Ordena uma lista usando Selection Sort.
    """
    for i in range(len(arr) - 1):
        min_idx = i  # Inicializa o índice do mínimo
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:  # Comparar com o menor encontrado até agora
                min_idx = j
        # Troca simultânea para evitar sobrescrever valores
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        print(f"Passo {i + 1}: lista = {arr}, índice mínimo = {min_idx}")
    return arr
```

```python
# B) 

import random

# Gerar lista aleatória de 7 números diferentes entre 1 e 50
lista = random.sample(range(1, 51), 7)
print("Lista original:", lista)

# Ordenar a lista usando Selection Sort corrigido
lista_ordenada = selection_sort(lista.copy())
print("Lista ordenada:", lista_ordenada)
```

```python
# C)

# Observando os prints da função selection_sort, percebemos como o índice mínimo evolui:

# - Em cada iteração 'i', o índice mínimo aponta para o menor elemento da sublista que começa no índice i.
# - No início, o índice mínimo geralmente está distante do índice i,
#   pois o menor valor pode estar em qualquer posição da lista.
# - Após a troca, a lista vai ficando mais ordenada, e nas próximas iterações
#   o índice mínimo tende a coincidir com o índice atual, eliminando trocas desnecessárias.

# Exemplo baseado na lista gerada:
# Se a lista fosse [42, 7, 19, 30, 11, 50, 25]:
# 1. i=0: menor é 7 (índice 1) → troca com 42 → lista: [7, 42, 19, 30, 11, 50, 25]
# 2. i=1: menor é 11 (índice 4) → troca com 42 → lista: [7, 11, 19, 30, 42, 50, 25]
# 3. i=2: menor é 19 (índice 2) → já no lugar correto
# 4. i=3: menor é 25 (índice 6) → troca com 30 → lista: [7, 11, 19, 25, 42, 50, 30]
# 5. i=4: menor é 30 (índice 6) → troca com 42 → lista: [7, 11, 19, 25, 30, 50, 42]
# 6. i=5: menor é 42 (índice 6) → troca com 50 → lista final: [7, 11, 19, 25, 30, 42, 50]

# Conclusão:
# A cada passo, o índice mínimo é atualizado para o menor valor da sublista restante,
# e podemos observar claramente sua evolução conforme a lista vai sendo ordenada.
```

```python
# D)

def find_max(lista):
    maior = lista[0]
    for n in lista:
        if n > maior:
            maior = n
    return maior

print("Maior elemento:", find_max(lista_ordenada))
```
